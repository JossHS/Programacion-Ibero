```python
#SUMA DE DOS NÚMEROS IF-ELSE

a = int(input("Ingresa el primer número: "))
b = int(input("Ingresa el segundo número: "))

c = a * b

if a * b == 100:
    print("Es correcto!", c)
else:
    print(f"La multiplicación no es 100, es: {c}!")
```
