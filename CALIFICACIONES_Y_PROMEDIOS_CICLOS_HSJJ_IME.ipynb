```python
#PEDIR AL USUARIO CUÁNTAS CALIFICACIONES QUIERE CAPTURAR

calif_1 = int(input("Ingresa la cantidad de calificaciones a capturar: "))

if calif_1 > 1 and calif_1 < 10:
    print(f"Las materias capturadas son: {calif_1}")
sum_t = 0
#Promediar calificaciones

for i in range(calif_1):
    califs = float(input(f"Ingresa la calificación {i+1}: "))
    sum_t += califs
prom = sum_t / calif_1
print(f"El promedio es: {round(prom, 2)}")



```
