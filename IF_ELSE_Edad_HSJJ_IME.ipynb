```python
#Código usando condicionales IF-ELSE
#José de Jesús Hernández Sánchez IME
#LINK GITHUB REPO: https://github.com/JossHS/Programacion-Ibero.git

dn = int(input("Ingresa tu día de nacimiento: "))
mn = int(input("Ingresa tu mes de nacimiento: "))
an = int(input("Ingresa tu año de nacimiento: "))

da = int(input("Ingresa el día actual: "))
ma = int(input("Ingresa el mes actual: "))
aa = int(input("Ingresa el año actual: "))

edad = aa - an

if ma > mn:
    edad = edad
elif ma < mn:
    edad = edad - 1
else:
    if da >= dn:
        edad = edad
    else:
        edad = edad - 1
print(f"Tu edad es: {edad}")



```
